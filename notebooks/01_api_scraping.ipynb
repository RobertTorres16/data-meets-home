# Scraping de la API de Idealista

Este notebook obtiene un token de acceso de la API de Idealista y descarga listados de propiedades en venta en Valencia.

> **Nota:** Sustituye `TU_API_KEY` y `TU_API_SECRET` por tus credenciales reales antes de ejecutar.

## 1. Autenticación — obtención del token de acceso

```python
import requests

# Credenciales de la API de Idealista
api_key = "TU_API_KEY"
api_secret = "TU_API_SECRET"

# URL de autenticación
auth_url = "https://api.idealista.com/oauth/token"

# Solicitar el token de acceso
response = requests.post(auth_url, auth=(api_key, api_secret), data={'grant_type': 'client_credentials'})

if response.status_code == 200:
    token = response.json()['access_token']
    print("Token de acceso obtenido correctamente.")
else:
    raise Exception(f"Error al obtener el token de acceso: {response.status_code} - {response.text}")
```

## 2. Búsqueda de propiedades en Valencia

```python
search_url = "https://api.idealista.com/3.5/es/search"

params = {
    'country': 'es',
    'operation': 'sale',
    'propertyType': 'homes',
    'center': '39.469907,-0.376288',  # Coordenadas de Valencia
    'distance': '10000',
    'maxItems': '50',
    'numPage': '1'
}

headers = {
    'Authorization': f'Bearer {token}',
    'Content-Type': 'application/x-www-form-urlencoded'
}

search_response = requests.post(search_url, headers=headers, data=params)

if search_response.status_code == 200:
    search_results = search_response.json()
    print("Resultados de la búsqueda obtenidos correctamente.")
else:
    raise Exception(f"Error al realizar la búsqueda: {search_response.status_code} - {search_response.text}")
```

## 3. Extracción y guardado de los datos

```python
import pandas as pd

propiedades = search_results.get('elementList', [])
print(f"Propiedades obtenidas en esta página: {len(propiedades)}")

# Convertir a DataFrame y guardar
df = pd.DataFrame(propiedades)
df.to_excel('../data/propiedades_valencia_2025.xlsx', index=False)
print("Datos guardados en ../data/propiedades_valencia_2025.xlsx")
```
